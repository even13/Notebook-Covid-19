```python
import pandas as pd
%matplotlib inline
import matplotlib.pyplot as plt 
```

```python
df=pd.read_csv('COVID-19-geographic-disbtribution-worldwide.csv', index_col=0)
```

```python
df
```

```python
subset=df.loc[['10/04/2020'],['countriesAndTerritories', 'cases', 'deaths']]
```

```python
subset.head()
```

```python
srtsub=subset.sort_values('cases',ascending=False)
```

```python
x=srtsub['countriesAndTerritories'].values.tolist()
#df.loc[['10/04/2020'], ['countriesAndTerritories']]
```

```python
y=srtsub['cases'].values.tolist()
```

```python
from matplotlib.pyplot import figure
from matplotlib.widgets import Slider, Button, RadioButtons
figure(num=None, figsize=(16, 6), dpi=80, facecolor='w', edgecolor='k')
plt.xticks(rotation=90, fontsize=6)
plt.plot(x,y)
```
